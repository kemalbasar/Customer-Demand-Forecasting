```python
#%%
df.head()
```

```python
px.scatter(df, x='GRANDTOTAL', y='Period')
```

```python
px.scatter(data_frame=df, x='Period', y='COUNTRY', size='GRANDTOTAL')
```

```python
px.scatter(data_frame=df,
           x='Period',
           y='COUNTRY',
           size='GRANDTOTAL'
           hover_name='COUNTRY',
           color='COUNTRY')
```

```python
px.scatter(data_frame=df,
           x='Period',
           y='GRANDTOTAL',
           size='SPRICE',
           hover_name='COUNTRY',
           color='COUNTRY',
           animation_frame='year',
           animation_group='COUNTRY')
```

```python
px.scatter(data_frame=df,
           x='gdpPercap',
           y='lifeExp',
           size='pop',
           hover_name='country',
           color='country',
           animation_frame='year',
           animation_group='country',
           range_x=[0,50000],
           range_y=[20,90])
```
